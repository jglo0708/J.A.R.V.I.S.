```python
from duplicate_finder import DuplicateFinder 
```

```python
dirName = 'Dataset Original/train'

finder = DuplicateFinder(dirName)
duplicate_pairs = finder.find_duplicates()
```

```python
img_list = []

for pair in duplicate_pairs:
    for img in pair:
        img_list.append(img)
```

```python
dirDuplicates = 'Dataset Original - Potential Duplicates'
dirName = 'Dataset Original/train'

if not os.path.exists(dirDuplicates):
    os.mkdir(dirDuplicates)
    print("Directory", dirDuplicates, "Created")
else:    
    print("Directory", dirDuplicates, "Already Exists")
    shutil.rmtree(dirDuplicates)
    os.mkdir(dirDuplicates)
    print("New Directory", dirDuplicates, "Created")


for img in img_list:
    pathname = os.path.join(dirName, img)
    shutil.copy2(pathname, dirDuplicates)

print('Duplicate Images Copied and Pasted')
            
```

```python
duplicates = dif.compare_images(dirDuplicates, similarity="normal", px_size=50, delete=False)
```

```python
df = pd.DataFrame(duplicates)
df.to_csv("duplicates_to_delete.csv", index=False)
```

```python
final_pairs = []

for pair in duplicate_pairs:
    if pair[0] in duplicates:
        final_pairs.append(pair)
    elif pair[1] in duplicates:
        final_pairs.append(pair)
```

```python
pairs_df = pd.DataFrame(final_pairs)
pairs_df.to_csv('duplicate_pairs.csv', index=False)
```

```python
img_list = []

for pair in final_pairs:
    for img in pair:
        img_list.append(img)

dirDuplicates = 'Dataset Original - Duplicates'
dirName = 'Dataset Original/train'

if not os.path.exists(dirDuplicates):
    os.mkdir(dirDuplicates)
    print("Directory", dirDuplicates, "Created")
else:    
    print("Directory", dirDuplicates, "Already Exists")
    shutil.rmtree(dirDuplicates)
    os.mkdir(dirDuplicates)
    print("New Directory", dirDuplicates, "Created")


for img in img_list:
    pathname = os.path.join(dirName, img)
    shutil.copy2(pathname, dirDuplicates)

print('Duplicate Images Copied and Pasted')
```
